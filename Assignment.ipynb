```python
from sklearn.datasets import load_breast_cancer # Breast Cancer dataset 불러오기
data = load_breast_cancer()
data
```

```python
import pandas as pd

X, y = pd.DataFrame(data.data, columns=data.feature_names), data.target # Breaset Cancer 데이터를 X, y로 나누어 저장
X
```

```python
X[X.isnull().any(axis=1)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x = X.mean().index, y = X.mean().values)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.model_selection import train_test_split # data를 나눠주는 코드

# train data와 test data의 X, y를 나눠서 저장하는 코드
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Decision Tree 불러오기
from sklearn.metrics import classification_report # for model evaluation metrics
from sklearn.tree import DecisionTreeClassifier # for decision tree models

# Decision Tree 분류기 불러오기
clf = DecisionTreeClassifier()
# Decision Tree 학습을 위해 parameter 채우기
clf.fit(X_train, y_train)
# Decision Tree 테스트를 위해 parameter 채우기
pred = clf.predict(X_test)
# accuracy 계산
acc = clf.score(X_test, y_test)

print("Accuracy : %.3f" % acc)
```

```python
# Random Forest 불러오기 
from sklearn.ensemble import RandomForestRegressor
# Random Forest 분류기 불러오기
clf = RandomForestRegressor(n_estimators=20, random_state=0)
# Random Forest 학습을 위해 parameter 채우기
clf.fit(X_train, y_train)
# Random Forest 테스트를 위해 parameter 채우기
pred = clf.predict(X_test)
# accuracy 계산
acc = clf.score(X_test, y_test)
print("Accuracy : %.3f" % acc)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# code from 6번
fig, ax = plt.subplots(figsize=(12, 20))
# 6번에서 완성 코드를 기준으로 feature importance 불러오기
feature_importances = pd.Series(clf.feature_importances_, index=X.columns)

y_pos = np.arange(len(feature_importances))

ax.barh(y_pos, feature_importances, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(X.columns)
ax.invert_yaxis()
ax.set_xlabel('Performance')
ax.set_title('Feature Importance of Random Forest Classifier')

plt.show()
```
