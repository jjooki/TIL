```python
data.index = pd.to_datetime(data.Timestamp, unit='s')
data
```

```python
data = data.drop(['Timestamp'], axis=1)
data
```

```python
data = data.resample('D').mean()
data
```

```python
data = data.dropna()
```

```python
from sklearn.preprocessing import RobustScaler

rob_scaler = RobustScaler()

for col in data.columns:
    data[col] = rob_scaler.fit_transform(data[col].values.reshape(-1, 1))
```

```python
data
```

```python
train = data.iloc[:-365]
test = data.iloc[-365:]
```

```python
X_train = train.drop(columns=['Weighted_Price'])
y_train = train.Weighted_Price

X_test = test.drop(columns=['Weighted_Price'])
y_test = test.Weighted_Price
```

# ML with scikit-learn

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
rf = RandomForestRegressor(n_estimators=100, max_depth=10)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
rmse
```

```python
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
rmse
```

# DL with tensorflow.keras

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Model, layers, optimizers, datasets

class LSTM(Model):
    def __init__(self, units):
        super(LSTM, self).__init__()
        
        self.sequential = tf.keras.Sequential([
            layers.LSTM(units, activation='relu'),
            layers.BatchNormalization(),
            layers.Dense(1)
        ])

    def call(self, x):
        out = self.sequential(x)
        return out
```

```python
X_train = X_train.values
y_train = y_train.values

X_test = X_test.values
y_test = y_test.values
```

```python
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
```

```python
model = LSTM(10)
```

```python
lr = 1e-3

model.compile(optimizer=optimizers.Adam(learning_rate=lr),
              loss='mse')
```

```python
batch_size = 32
epochs = 10

model.fit(X_train, y_train, validation_split=0.2, batch_size=batch_size, epochs=epochs, shuffle=True, verbose=1)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)
mean_squared_error(y_test, y_pred, squared=False)
```
